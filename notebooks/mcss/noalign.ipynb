```python
X = []
for protein in proteins:
    prot = Protein(protein, stats['stats41'], paths)
    ligs = prot.lm.get_xdocked_ligands(20)
    for lig in ligs:
        X += [(protein, lig,
              prot.lm.path('LIGANDS', {'ligand': lig}),
              prot.lm.path('DOCK_PV', {'ligand': lig}))
              ]
        print(X[-1])
```

```python
import pandas as pd
df = pd.DataFrame(X, columns=['protein', 'ligand', 'reference_pose', 'docked'])
df['reference_pose'] = [x.replace('/Users/jpaggi/sherlock/oak/users/jpaggi/combind', '/oak/stanford/groups/rondror/users/jpaggi/bpp_data') for x in df['reference_pose']]
df['reference_pose'] = [x.replace('ligands', 'ligands/prepared_ligands') for x in df['reference_pose']]
df['docked'] = [x.replace('/Users/jpaggi/sherlock/oak/users/jpaggi/combind', '/oak/stanford/groups/rondror/users/jpaggi/bpp_data') for x in df['docked']]
df['ifp'] = ['/oak/stanford/groups/rondror/users/jpaggi/rdkit_combind_features/{}/ifp/{}.h5'.format(protein, ligand)
             for protein, ligand in zip(df['protein'], df['ligand'])]
df['ifp_scores'] = ['/oak/stanford/groups/rondror/users/jpaggi/rdkit_combind_features/{}/ifp/{}-scores.h5'.format(protein, ligand)
             for protein, ligand in zip(df['protein'], df['ligand'])]

df['rmsd'] = ['/oak/stanford/groups/rondror/users/jpaggi/rdkit_combind_features/{}/rmsd/{}.h5'.format(protein, ligand)
             for protein, ligand in zip(df['protein'], df['ligand'])]
df['dock_score'] = ['/oak/stanford/groups/rondror/users/jpaggi/rdkit_combind_features/{}/dock_score/{}.h5'.format(protein, ligand)
                    for protein, ligand in zip(df['protein'], df['ligand'])]

df.to_csv('data.csv', index=False)
df
```

```python
for protein in proteins:
    prot = Protein(protein, stats['stats104'], paths)
    xtal = prot.lm.st + '_crystal_lig'
    ligs = prot.lm.get_xdocked_ligands(20)

    print(protein, xtal, ligs)

    prot.load_docking([prot.lm.st + '_crystal_lig'], load_crystal=True)
    prot.load_docking(ligs, load_mcss=True)

    for lig in ligs:
        for i, pose in enumerate(prot.docking[prot.lm.st][lig].poses):
            if pose.rmsd < 2.0:
                print(lig, pose.rmsd, prot.lm.mcss.get_rmsd(xtal, lig, 0, i))
                break
```

```python
for protein in proteins:
    prot = Protein(protein, stats['stats104'], paths)
    ligs = prot.lm.get_xdocked_ligands(20)
    prot.load_docking(ligs, load_mcss=True)
    print(protein)
    for i, lig1 in enumerate(ligs):
        for lig2 in ligs[i+1:]:
            done=False
            for r1, pose1 in enumerate(prot.docking[prot.lm.st][lig1].poses):
                if pose1.rmsd > 2.0: continue
                for r2, pose2 in enumerate(prot.docking[prot.lm.st][lig2].poses):
                    if pose2.rmsd < 2.0:
                        if prot.lm.mcss.get_rmsd(lig1, lig2, r1, r2) and prot.lm.mcss.get_rmsd(lig1, lig2, r1, r2) > 3.0:
                            print(lig1, lig2, pose1.rmsd, pose2.rmsd, prot.lm.mcss.get_rmsd(lig1, lig2, r1, r2))
                        done = True
                        break
                if done:
                    break
    print()
```
